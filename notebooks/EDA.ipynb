## Looking for and Handling Missing values

```python
missing = df.isnull().sum().sort_values(ascending=False)
percent_missing = (df.isnull().mean() * 100).sort_values(ascending=False)
pd.concat([missing, percent_missing], axis=1, keys=['Missing Count', 'Missing %']).head(20)
```

```python
df.nunique().sort_values()
```

```python
df = df.replace("?", np.nan)
```

```python
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
df['readmitted'].value_counts()
```

```python
df["readmit_binary"] = df["readmitted"].apply(lambda x: 1 if x in ["<30"] else 0)
```

## <b><b> Graphs to see the distribution of data, compare categories

Data Distribution of readmission 

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='readmitted')
plt.title("Readmission Distribution")
plt.show()
```

Data Distribution of age

```python
plt.figure(figsize=(8,4))
sns.countplot(data=df, x='age', order=df['age'].value_counts().sort_index().index)
plt.xticks(rotation=45)
plt.title("Age Group Distribution")
plt.show()
```

Data Distribution of gender

```python
sns.countplot(data=df, x='gender')
plt.title("Gender Distribution")
plt.show()
```

```python
df['admission_type_id'].value_counts().head(10)
df['discharge_disposition_id'].value_counts().head(10)
```

Data Distribution of lab procedures

```python
sns.histplot(df['num_lab_procedures'], bins=40)
plt.title("Distribution of Lab Procedures")
plt.show()
```

Data Distribution of medications count

```python
sns.histplot(df['num_medications'], bins=40)
plt.title("Distribution of Medications Count")
plt.show()
```

Comparing the categories of age and readmission

```python
plt.figure(figsize=(8,4))
sns.countplot(data=df, x='age', hue='readmit_binary', order=df['age'].value_counts().sort_index().index)
plt.xticks(rotation=45)
plt.title("Age vs Readmission")
plt.show()
```

Comparing the categories of gender and readmission

```python
sns.countplot(data=df, x='gender', hue='readmit_binary')
plt.title("Gender vs Readmission")
plt.show()
```

Correaltion Heatmap of Numerical Features

```python
numeric_cols = df.select_dtypes(include=np.number)
plt.figure(figsize=(14,10))
sns.heatmap(numeric_cols.corr(), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

Comparing Medication Change and Readmission Status

```python
med_cols = ['change', 'diabetesMed']

for col in med_cols:
    print(f"\nValue counts for {col}:")
    print(df[col].value_counts())
    sns.countplot(data=df, x=col, hue='readmit_binary')
    plt.title(f"{col} vs Readmission")
    plt.show()
```

Distribution of Medical Specialty 

```python
plt.figure(figsize=(10,5))
df['medical_specialty'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Medical Specialties")
plt.show()
```
